# TP — Visualisation des fonctions d'activation (NumPy + Matplotlib)

**But :** implémenter et visualiser 4 fonctions d'activation (ReLU, LeakyReLU, Sigmoid, Tanh) avec **NumPy** et **Matplotlib**, puis produire :
1. un **tracé comparatif** (toutes les courbes sur une figure)
2. une **grille de subplots 2×2** (une courbe par sous-figure)

## Partie A — Mise en place

### A1) Imports et vecteur x

```python
import numpy as np
import matplotlib.pyplot as plt

# Vecteur x : 600 points entre -3 et 3
x = np.linspace(-3, 3, 600)
```

### A2) Dictionnaire des fonctions d'activation

```python
# Fonctions d'activation vectorisées (sans boucle Python)
activations = {
    "Relu":      lambda x: np.maximum(0, x),
    "Sigmoid":   lambda x: 1 / (1 + np.exp(-x)),
    "Tanh":      lambda x: np.tanh(x),
    "LeakyRelu": lambda x: np.where(x > 0, x, 0.1 * x),
}
```

### A3) Dictionnaires de style (couleurs + marqueurs)

```python
couleurs = {
    "Relu":      "tab:blue",
    "Sigmoid":   "tab:orange",
    "Tanh":      "tab:green",
    "LeakyRelu": "tab:red",
}

marques = {
    "Relu":      "o",
    "Sigmoid":   "s",
    "Tanh":      "^",
    "LeakyRelu": "D",
}
```

## Partie B — Tracé comparatif (1 seule figure)

```python
plt.figure(figsize=(8, 5))

for name, func in activations.items():
    plt.plot(
        x, func(x),
        label=name,
        color=couleurs[name],
        marker=marques[name],
        markersize=4,
        markevery=50
    )

plt.title("Comparaison des fonctions d'activation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("activations_all.png", dpi=150)
plt.show()
```

## Partie C — Subplots (grille 2×2)

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7), sharex=True, sharey=False)
axes = axes.ravel()

for i, (name, func) in enumerate(activations.items()):
    ax = axes[i]
    ax.plot(
        x, func(x),
        color=couleurs[name],
        marker=marques[name],
        markersize=4,
        markevery=50,
        label=name
    )
    ax.set_title(name)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)
    ax.legend()

plt.suptitle("Fonctions d'activation — Grille 2×2", fontsize=13)
plt.tight_layout()
plt.savefig("activations_subplots.png", dpi=150)
plt.show()
```

## Extension — Softplus et ELU

Ajout de deux fonctions supplémentaires :
- **Softplus** : $\text{softplus}(x) = \ln(1 + e^x)$
- **ELU** ($\alpha = 1$) : $\text{elu}(x) = x$ si $x > 0$, sinon $\alpha(e^x - 1)$

```python
# Ajout de Softplus et ELU
activations["Softplus"] = lambda x: np.log(1 + np.exp(x))
activations["ELU"]      = lambda x: np.where(x > 0, x, 1.0 * (np.exp(x) - 1))

couleurs["Softplus"] = "tab:purple"
couleurs["ELU"]      = "tab:brown"

marques["Softplus"] = "*"
marques["ELU"]      = "x"
```

```python
# Tracé comparatif avec extension
plt.figure(figsize=(9, 5))

for name, func in activations.items():
    plt.plot(
        x, func(x),
        label=name,
        color=couleurs[name],
        marker=marques[name],
        markersize=4,
        markevery=50
    )

plt.title("Comparaison des fonctions d'activation (avec extension)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("activations_all_extension.png", dpi=150)
plt.show()
```

```python
# Subplots avec extension (grille 2×3)
fig, axes = plt.subplots(2, 3, figsize=(14, 7), sharex=True)
axes = axes.ravel()

for i, (name, func) in enumerate(activations.items()):
    ax = axes[i]
    ax.plot(
        x, func(x),
        color=couleurs[name],
        marker=marques[name],
        markersize=4,
        markevery=50,
        label=name
    )
    ax.set_title(name)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)
    ax.legend()

plt.suptitle("Fonctions d'activation — Extension (2×3)", fontsize=13)
plt.tight_layout()
plt.savefig("activations_subplots_extension.png", dpi=150)
plt.show()
```
